```python
rocks = ['quartz', 'granite', 'diamond', 'sandstone']
for position in range(len(rocks)):
    if rocks[position] == 'diamond':
        print("found diamond at position:", position)
        break
```

```python
# else at the end of a for loop
rocks = ['quartz', 'granite', 'sandstone']
for position in range(len(rocks)):
    if rocks[position] == 'diamond':
        print("found diamond at position:", position)
        break
else:
    print("no diamond found")
```

```python
for filename in sorted(glob.glob('data/inflammation*.csv')):
    data = numpy.loadtxt(fname=filename, delimiter=',')
    print('Examining:', filename)
    if data.max(axis=0)[0] == 0 and data.max(axis=0)[20] == 20:
        print('Suspicious looking maxima!')
    elif data.min(axis=0).sum() == 0:
        print('Minima add up to zero!')
    else:
        print('Seems OK')
```
